# Thuật toán Ứng dụng - Phase 2: Xử lý dữ liệu

Tính số lượng giá trị NaN - không xác định

```python
df.isna().sum()
```

Để dễ nhìn hơn thì sẽ visualize dưới dạng bar chart sử dụng thư viện `missingno`

```python
import missingno as msno

msno.bar(df)
plt.show()
```

Có thể thấy `collision_type`, `property_damage`, `police_report_available` là 3 trường có dấu hiệu thiếu. Nhận thấy đây là một trong các trường có thể được sử dụng để xác định hành vi lừa đảo, mình sẽ k remove mà fill **giá trị xuất hiện nhiều nhất** trong cột đấy thay thế cho giá trị khuyết.

```python
df['collision_type'] = df['collision_type'].fillna(df['collision_type'].mode()[0])
df['property_damage'] = df['property_damage'].fillna(df['property_damage'].mode()[0])
df['police_report_available'] = df['police_report_available'].fillna(df['police_report_available'].mode()[0])
df['authorities_contacted']  = df['authorities_contacted'].fillna(df['authorities_contacted'].mode()[0])
# Thực hiện tính toán N/A lại để kiểm tra
df.isna().sum()
```

```python
## Thuật toán Ứng dụng - Phase 3: Biểu diễn dữ liệu
```

The Distplot depicts data by a histogram and a line in combination to it. While the histogram plot the distribution of the values, the smooth line 
indicates the estimated probability density function of the variable using Kernel density estimate (Kde) plot.

Take the month_as_customer for example, we can easily see that height of columns between 200 and 300 are the highest which means that most of the data for this feature lies on this range

```python
plt.figure(figsize=(25, 20))
plot_number = 1

for col in X.columns:  
    if plot_number <= 24:
        ax = plt.subplot(5, 5, plot_number)
        sns.distplot(X[col])
        plt.xlabel(col, fontsize=15)
    
    plot_number += 1

plt.tight_layout()
plt.show()
```

Outliers detection using boxplot() function

- Box plot is a data visualization plotting function. It shows the min, max, median, first quartile, and third quartile. The dots are the outliers.
Let's take month_as_customer figure as an example again. It can easily be seen min of the feature is 0, while max is 500. In addition, the red rectangle in the middle is the interquatile range which runs from first quartile 25% to third quartile 75% while the line in the middle is the median of the feature

```python
plt.figure(figsize=(20, 15))
plot_number = 1

for col in X.columns:  
    if plot_number <= 24:
        ax = plt.subplot(5, 5, plot_number)
        sns.boxplot(X[col], orient = 'h')
    
    plot_number += 1

plt.tight_layout()
plt.show()
```

```python
#splitting data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
from sklearn.ensemble import RandomForestClassifier

rand_clf = RandomForestClassifier(n_estimators = 75, criterion='entropy', random_state = 50)
rand_clf.fit(X_train, y_train)

y_pred = rand_clf.predict(X_test)
```

n_estimators: the number of decision trees in the forest

criterion: specify the function to measure a quality of a split. "entropy"(information gain criterion) used due to it advantages in handling class imbalance

random_state: random number of generation during the training process

```python
# accuracy_score, confusion_matrix and classification_report

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

rand_clf_train_acc = accuracy_score(y_train, rand_clf.predict(X_train))
rand_clf_test_acc = accuracy_score(y_test, y_pred)

print(f"Training accuracy of Random Forest is : {rand_clf_train_acc}")
print(f"Test accuracy of Random Forest is : {rand_clf_test_acc}")

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

Training accuracy is 100%, however test accuracy is 80.4% which indicates that the model has learned the training data perfectly, possibly overfitting.
Therefore, in order to increase the precision of result, we could conduct hyperparmeter tuning or balance the dataset.
